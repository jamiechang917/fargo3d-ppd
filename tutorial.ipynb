```python
from fargo3d.data import Data
from fargo3d.visual import *
import astropy.units as u

# dark mode
plt.style.use('dark_background')
plt.rcParams['axes.edgecolor'] = '#FAFBFCCE'
plt.rcParams['axes.labelcolor'] = '#FAFBFCCE'
plt.rcParams['xtick.color'] = '#FAFBFCCE'
plt.rcParams['ytick.color'] = '#FAFBFCCE'

path = r"C:\Users\jamiechang917\Desktop\fargo3d-data\data"
snap = 1000
n_dust = 1
unit_length =  100.0*u.au
unit_mass = 1.0*u.M_sun
data = Data(path=path, snap=snap, n_dust=n_dust, unit_length=unit_length, unit_mass=unit_mass)
```

```python
plot_velocity_map(data=data, r_bin=2, z_bin=6, scale=6,
                  phi_min=-np.pi, phi_max=np.pi,
                  # phi_min=-0.5, phi_max=0.5,
                  r_min=0, r_max=200,
                  z_min=0.0, z_max=50,
                  fluid='gas', coordinate='cyl')
```
